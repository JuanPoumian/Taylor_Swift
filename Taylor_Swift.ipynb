# Analyze Taylor Swift Lyrics

```python
# Importing the libraries needed for sentiment analysis
%matplotlib inline

import pandas as pd
import string
import seaborn as sns
import matplotlib.pyplot as plt
import collections
import nltk
from nltk.tokenize import word_tokenize
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
# Load the dataset containing Taylor Swift lyrics
lyrics = pd.read_csv("taylor_swift_lyrics_2006-2022_all.csv")

# Inspect the first few rows of the dataset
lyrics.head()
```

```python
# Get information about the DataFrame, such as the number of rows and columns, and data types of columns
lyrics.info()
```

```python
# Get a list of all the unique album names in the dataset
print(lyrics.album_name.unique())
```

```python
# This function maps the name of each album to the year it was released
def album_release(row):  
    if row['album_name'] == 'Taylor Swift':
        return '2006'
    elif row['album_name'] == 'Fearless (Taylor’s Version)':
        return '2008'
    elif row['album_name'] == 'Speak Now (Deluxe)':
        return '2010'
    elif row['album_name'] == 'Red (Deluxe Edition)':
        return '2012'
    elif row['album_name'] == '1989 (Deluxe)':
        return '2014'
    elif row['album_name'] == 'reputation':
        return '2017'
    elif row['album_name'] == 'Lover':
        return '2019'
    elif row['album_name'] == 'evermore (deluxe version)':
        return '2020'
    elif row['album_name'] == 'folklore (deluxe version)':
        return '2021'
    # This album name is recorded two ways, so the function checks if 'midnights' is in the album name
    elif 'midnights' in row['album_name']:
        return '2022'

# Apply the function to each row of the DataFrame to create a new column containing the year each album was released
lyrics['album_year'] = lyrics.apply(lambda row: album_release(row), axis=1)

# Inspect the first few rows of the DataFrame to verify the new column was added successfully
lyrics.head()# Convert all lyrics to lowercase
lyrics['clean_lyric'] = lyrics['lyric'].str.lower()

# Remove all punctuation from the lyrics
lyrics['clean_lyric'] = lyrics['clean_lyric'].str.replace('[^\w\s]','', regex=True)

# Inspect the first few rows of the DataFrame to verify the new column was added successfully
lyrics.head()
```

```python
# Convert all lyrics to lowercase
lyrics['clean_lyric'] = lyrics['lyric'].str.lower()

# Remove all punctuation from the lyrics
lyrics['clean_lyric'] = lyrics['clean_lyric'].str.replace('[^\w\s]','', regex=True)

# Inspect the first few rows of the DataFrame to verify the new column was added successfully
lyrics.head()
```

```python
# Define a list of stopwords that we want to remove from the lyrics
stop = ['the', 'a', 'this', 'that', 'to', 'is', 'am', 'was', 'were', 'be', 'being', 'been']

# For each lyric in our dataset, remove all stopwords by first splitting the lyric into words,
# then removing all the words in our list of stopwords,
# and finally joining the words back together into a string
lyrics['clean_lyric'] = lyrics['clean_lyric'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))

# Inspect the first few rows of the DataFrame to verify that the stopwords have been removed
lyrics.head()
```

```python
# Create a new column that indicates whether each song's lyrics contain the word "midnight"
lyrics['midnight'] = lyrics['clean_lyric'].str.contains('midnight')

# Count the number of times "midnight" appears in the lyrics across all songs
sum(lyrics['midnight'])
```

```python
# Define lists of time-related words that we want to search for in the lyrics
night = ['night','midnight', 'dawn', 'dusk', 'evening', 'late', 'dark', '1am', '2am', '3am', '4am']
day = ['day', 'morning', 'light', 'sun', 'dawn', 'noon', 'golden', 'bright']
time = ['today', 'tomorrow', 'yesterday']

# Create a regular expression string for each list of words
night_regex = '|'.join(night)
day_regex = '|'.join(day)
time_regex = '|'.join(time)

# Create new columns in the DataFrame for each category of words, using regular expression matching
lyrics['night'] = lyrics['clean_lyric'].str.contains(night_regex)
lyrics['day'] = lyrics['clean_lyric'].str.contains(day_regex)
lyrics['time'] = lyrics['clean_lyric'].str.contains(time_regex)

# Count the number of times each category of word appears in the lyrics
night_count = sum(lyrics['night'])
day_count = sum(lyrics['day'])
time_count = sum(lyrics['time'])

# Print the count of each category of word
print("night words: ", night_count)
print("day words: ", day_count)
print("time words: ", time_count)

# Inspect the first few rows of the DataFrame to verify that the new columns have been added
lyrics.head()
```

```python
# Create a new dataframe to group mentions by year
yearly_mentions = lyrics.groupby('album_year').sum().reset_index()

# Reset the index to make the year a column instead of an index
# and aggregate the mentions for each year using the sum function
yearly_mentions
```

```python
# Plot the mentions of the word "night" over the years
plt.plot(yearly_mentions['album_year'], yearly_mentions['night'], color='red')

# Set the title of the plot
plt.title("Taylor Swift Night Mentions")

# Show the plot
plt.show()
```

```python
# Read in the file that contains album names and their respective release years
year_name = pd.read_csv('album_year_name.csv')

# Sort both dataframes by year in ascending order
yearly_mentions.sort_values(by='album_year', ascending=True, inplace=True)
year_name.sort_values(by='album_year', ascending=True, inplace=True)

# Add a new column to the yearly_mentions dataframe that contains the album names
yearly_mentions['album_name'] = year_name['album_name']
```

```python
# Sort the yearly mentions DataFrame by the 'night' column in descending order to find the albums with the most night references
yearly_mentions.sort_values(by='night', ascending=False)
```

```python
# Sort the yearly_mentions DataFrame by the 'day' column in descending order to find the albums with the most day references
yearly_mentions.sort_values(by='day', ascending=False)
```

```python
# Create a plot with two lines: one showing the number of night references by year and another showing the number of day references by year
plt.plot(yearly_mentions['album_year'], yearly_mentions['night'], label='night', color='purple')
plt.plot(yearly_mentions['album_year'], yearly_mentions['day'], label='day', color='orange')

# Add a title and legend to the plot
plt.title("Taylor Swift Day vs. Night Mentions")
plt.legend()

# Display the plot
plt.show()
```

```python
# Create a position variable that combines the track number and line number to create a unique position for each line of lyrics
lyrics['position'] = lyrics['track_n'] + (lyrics['line']/1000)

# Create a new DataFrame that is grouped by position and shows the total number of day and night references at each position
positional_mentions = lyrics.groupby('position').sum().reset_index()

# Increase the size of the plot to make it more readable
fig = plt.gcf()
fig.set_size_inches(25,10)

# Create a plot with two lines to show the frequency of day vs. night references at each position in the album
plt.plot(positional_mentions['position'], positional_mentions['night'], label='night', color='purple')
plt.plot(positional_mentions['position'], positional_mentions['day'], label='day', color='orange')

# Add a legend and title to the plot
plt.legend()
plt.title("Day vs. Night Mentions by Album Position")

# Display the plot
plt.show()
```

```python
# Tokenize the words in the 'clean_lyric' column by splitting each lyric string on whitespace
lyrics['lyrics_tok'] = lyrics['clean_lyric'].str.split(' ')
```

```python
# Download the "vader_lexicon" package from the Natural Language Toolkit (NLTK) for use in our sentiment analyzer.
nltk.download('vader_lexicon')
```

```python
# Create a SentimentIntensityAnalyzer object from the NLTK package and use it to analyze the sentiment of the given text.
sia = SentimentIntensityAnalyzer()
sia.polarity_scores("I love Taylor Swift!")
```

```python
# Create a new column called "polarity" in the "lyrics" DataFrame and apply the SentimentIntensityAnalyzer's polarity_scores method to the "clean_lyric" column using a lambda expression.
lyrics['polarity'] = lyrics['clean_lyric'].apply(lambda x: sia.polarity_scores(x))

# Display the first few rows of the DataFrame.
lyrics.head()
```

```python
# Extract the values from the "polarity" dictionary in the "lyrics" DataFrame and create new columns for them, named 'neg', 'neu', 'pos', and 'compound'.
lyrics[['neg', 'neu', 'pos', 'compound']] = lyrics['polarity'].apply(pd.Series)

# Drop the "polarity" column from the DataFrame since it is no longer needed.
lyrics.drop('polarity', axis=1)
```

```python
# Calculate the overall sentiment scores for positive, negative, and compound by summing the corresponding columns in the "lyrics" DataFrame.
pos = sum(lyrics['pos'])
neg = sum(lyrics['neg'])
compound = sum(lyrics['compound'])

# Display the overall sentiment scores.
print("positive: ", pos)
print("negative: ", neg)
print("compound: ", compound)
# This means that, overall, the lyrics tend to be more positive than negative, but the sentiment leans towards neutral.
```

```python
# Group the "lyrics" DataFrame by album year using the groupby method, calculate the sum of sentiment scores for each year using the sum method, and create a new DataFrame with the results using the reset_index method.
yearly_sentiment = lyrics.groupby('album_year').sum().reset_index()

# Plot the compound sentiment score over time using the matplotlib library.
plt.plot(yearly_sentiment['album_year'], yearly_sentiment['compound'], color='brown')
plt.title("Taylor Swift's average album sentiment")
plt.show()
```
